```python
import numpy as np
import pandas as pd
```

```python
craft = pd.read_csv('amazon.csv')
```

```python
craft.head()
```

```python
product_craft =  craft[['product_id','product_name','category','actual_price','rating','rating_count','about_product','img_link']].copy()
```

```python
product_craft.sample(3)
```

```python
product_craft['rating_count'] = product_craft['rating_count'].str.replace(',','')
```

```python
product_craft.sample(3)
```

```python
product_craft['rating_count'] = product_craft['rating_count'].fillna(0)
product_craft['rating_count'] = product_craft['rating_count'].astype(int)
product_craft['rating_count'] = product_craft['rating_count']/1000
```

```python
product_craft.sample(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(x = "rating_count" ,data = product_craft)

plt.title("Rating Graph")
plt.xlabel("rating")
plt.ylabel("rating_count")
plt.show()
```
